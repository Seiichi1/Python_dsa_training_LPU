GCD using Euclidean algorithm

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("GCD of", a, "and", b, "is:", gcd(a, b))
```

```python
a=12
b=18

lcm = (a * b) // gcd(a, b)
print("LCM of", a, "and", b, "is:", lcm)
```

```python
#star pyramid
n=5
for i in range(0,n):
    print(" "*(n-i-1),end="")
    print(" *"*(i+1))
```

```python
#hollow square
n=5
for i in range(n):
    if i==0 or i==n-1:
        print("*"*n)
    else:
        print("*"*1,end="")
        print(" "*(n-2),end="")
        print("*"*1)
```

```python
#floyd triangle
n=15
l=[i*(i+1)//2 for i in range(n)]
print(l)
for i in range(1,n):
    print(i,end=" ")
    if i in l:
        print()
    
    
```
