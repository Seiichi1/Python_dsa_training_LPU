GCD using Euclidean algorithm

```python
def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("GCD of", a, "and", b, "is:", gcd(a, b))
```

```python
a=12
b=18

lcm = (a * b) // gcd(a, b)
print("LCM of", a, "and", b, "is:", lcm)
```
