Given an integer n, return the number of prime numbers that are strictly less than n.

 

Example 1:

Input: n = 10
Output: 4
Explanation: There are 4 prime numbers less than 10, they are 2, 3, 5, 7.

```python
#1st Solution
n=10
prime=0
def is_prime(n): 
    if n<=1:
        return False   
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
for i in range(2,n):
    if is_prime(i):
        prime+=1
print(prime)
```

```python
#2nd Solution using Sieve of Eratosthenes
n=10
if n<=1:
    print(0)
is_prime=[True]*n
is_prime[0]=is_prime[1]=False
for p in range(2,int(n**0.5)+1):
    if is_prime[p]:
        is_prime[p*p:n:p]=[False]*len(range(p*p,n,p))
print(sum(is_prime))
```
