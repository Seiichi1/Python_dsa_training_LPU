```python
#difference between for and for each

x=[1, 2, 3, 4, 5]

#for
for i in range(len(x)):
    print(i)

print()
# for each
for element in x:
    print(element)
```

```python
#print index and value of list using enumerate

x=[10, 20, 30, 40, 50]
for index, value in enumerate(x):
    print(f"Index: {index}, Value: {value}")
```

```python
#difference between in-place and out-of-place algorithms
x=[1, 2, 3, 4, 5]
#in-place
y=x.sort()
print(x)    
print(y) #None because sort() method modifies the list in place and returns None

#out-of-place
z=sorted(x)
print(z) #sorted() function returns a new sorted list without modifying the original list
print(x) #original list remains unchanged
```

```python
#defaultdict is a subclass of the built-in dict class. used to provide default values for keys that are not present in the dictionary. It takes a default factory function as an argument, which is called to provide a default value when a key is accessed that does not exist in the dictionary.
from collections import defaultdict

res=defaultdict(list)
res['a'].append(1)
print(res)
```
